```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
```

```python
html = urlopen('https://www.bloomberg.com/quote/PG:US')

bsobj=BeautifulSoup(html.read(), 'lxml')
```

```python
print(bsobj)
```

```python
nameList = bsobj.findAll('div',{'class':'headline__07dbac92'})
articlelist = []
for name in nameList:
    articlelist.append(name.get_text())
```

```python
print(articlelist)
```

```python
clean_articlelist = []
for i in articlelist:
    i = i.replace("’", "")
    i = i.replace("‘", "")
    clean_articlelist.append(i)
```

```python
clean_articlelist
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(clean_articlelist)
```

```python
df
```

```python
df.to_csv('PG_news.csv')
```
